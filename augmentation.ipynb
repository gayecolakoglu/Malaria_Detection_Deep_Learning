```python
def augment_using_ops(path):
  import matplotlib.pyplot as plt
  for filename in os.listdir(path):
    try:
      print(filename)
      print(os.path.join(path,filename))
      img = cv2.imread(os.path.join(path,filename))
      img = tf.image.random_flip_left_right(img)
      img = tf.image.random_flip_up_down(img)
      img = tf.image.rot90(img)
      img = np.array(img)
      plt.imshow(img)
      cv2.imwrite(path+'augmented'+filename,img)

    except:
      continue
```

```python
path = "/content/gdrive/MyDrive/Colab Notebooks/ML/CLAHE/Parasitized/"
augment_using_ops(path)
```

```python
path2 = "/content/gdrive/MyDrive/Colab Notebooks/ML/CLAHE/Uninfected/"
augment_using_ops(path2)
```
